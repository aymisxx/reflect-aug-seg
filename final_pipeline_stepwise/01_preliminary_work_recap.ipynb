## Scientific Motivation

LiDAR intensity is useful, but it is not a clean intrinsic surface descriptor.

For a point
$$
p_i = (x_i, y_i, z_i, I_i)
$$
the Euclidean range is
$$
R_i = \sqrt{x_i^2 + y_i^2 + z_i^2}
$$

A simplified LiDAR return interpretation is:
$$
I_i \propto \frac{\rho_i \cos(\alpha_i)}{R_i^2}
$$
where:
- $\rho_i$ is surface reflectivity,
- $\alpha_i$ is incidence angle,
- $R_i$ is range.

This means raw intensity mixes together:
- surface behavior,
- distance effects,
- measurement geometry,
- and sensor-specific effects.

So the project does **not** treat raw intensity as reflectivity.

Instead, the preliminary work introduced a practical proxy:
$$
\hat{\rho}_i = I_i \cdot R_i
$$

This is **not calibrated physical reflectivity**.
It is a **heuristic range-aware feature** whose value lies in semantic interpretability and lightweight signal augmentation, not in exact reflectance recovery.

That honesty clause remains fully active in this notebook and throughout the final project.

## Locate the dataset and verify the sequence structure

We define the SemanticKITTI sequence path and check whether the expected

```python
from pathlib import Path

# This recap notebook lives in final_pipeline/, so we allow both:
# 1) the original preliminary path
# 2) the planned final-stage renamed path
candidate_roots = [
    Path("../data/semantickitti_subset/dataset/sequences/00"),
    Path("../data/semantickitti/dataset/sequences/00"),
]

DATASET_ROOT = next((p for p in candidate_roots if p.exists()), candidate_roots[0])

expected_items = [
    "velodyne",
    "labels",
    "calib.txt",
    "poses.txt",
    "times.txt",
]

print("Dataset root:", DATASET_ROOT.resolve())
print("Exists:", DATASET_ROOT.exists())
print()

for item in expected_items:
    item_path = DATASET_ROOT / item
    print(f"{item:10} -> exists: {item_path.exists()} | path: {item_path}")

print()
if DATASET_ROOT.exists():
    print("Contents of sequence folder:")
    for p in sorted(DATASET_ROOT.iterdir()):
        print(" -", p.name)
else:
    print("Dataset root not found. Checked these candidate paths:")
    for p in candidate_roots:
        print(" -", p.resolve())
```

## List available frames and confirm file counts

We inspect the `velodyne` and `labels` folders to confirm that frame
files are present and that both folders contain matching counts.

This gives us a quick structural sanity check before loading the first frame.

If the counts and frame IDs match, then the dataset structure is not only present,
but internally consistent enough for frame-level loading in the next combo.

```python
velodyne_dir = DATASET_ROOT / "velodyne"
labels_dir = DATASET_ROOT / "labels"

velodyne_files = sorted(velodyne_dir.glob("*.bin"))
label_files = sorted(labels_dir.glob("*.label"))

print("Number of .bin frames   :", len(velodyne_files))
print("Number of .label frames :", len(label_files))
print()

if velodyne_files:
    print("First 5 velodyne frames:")
    for f in velodyne_files[:5]:
        print(" -", f.name)

print()

if label_files:
    print("First 5 label frames:")
    for f in label_files[:5]:
        print(" -", f.name)

print()
if velodyne_files and label_files:
    velodyne_stems = [f.stem for f in velodyne_files]
    label_stems = [f.stem for f in label_files]

    print("Counts match   :", len(velodyne_files) == len(label_files))
    print("Frame IDs match:", velodyne_stems == label_stems)

    if velodyne_stems == label_stems:
        print("Sample frame ID:", velodyne_stems[0])
```

## Load the first LiDAR frame

We load the first `.bin` file from the `velodyne` folder.

Each SemanticKITTI LiDAR frame stores raw point-cloud data as `float32` values
arranged as four channels per point:

- `x`
- `y`
- `z`
- `intensity`

The goal here is to load the first frame, reshape it into an `N x 4` array, and
inspect its basic structure.

This is the first actual contact with the raw LiDAR data.

```python
import numpy as np

frame_id = velodyne_files[0].stem
bin_path = velodyne_dir / f"{frame_id}.bin"

points = np.fromfile(bin_path, dtype=np.float32).reshape(-1, 4)

print("Frame ID:", frame_id)
print("File path:", bin_path)
print("Points array shape:", points.shape)
print("Data type:", points.dtype)
print()

print("First 5 points:")
print(points[:5])
```

## Load the matching label frame

We load the label file corresponding to the same frame.

```python
label_path = labels_dir / f"{frame_id}.label"

labels = np.fromfile(label_path, dtype=np.uint32)

print("Frame ID:", frame_id)
print("File path:", label_path)
print("Labels array shape:", labels.shape)
print("Data type:", labels.dtype)
print()

print("First 10 raw labels:")
print(labels[:10])
```

## Verify point-label alignment

## Define simple range-aware pseudo-reflectivity features

We construct simple reflectivity-style proxy signals from the verified
single frame.

These are not physically calibrated reflectivity values.
They are heuristic range-aware transformations of raw intensity that let us inspect whether
compensating for distance reveals a more interpretable signal structure.

We begin with two simple variants:

$$
\hat{\rho}^{(1)} = I \cdot R
$$

and

$$
\hat{\rho}^{(2)} = I \cdot R^2
$$

The purpose of defining both is not to claim that both are good.
It is to compare a moderate range-aware transformation against a stronger one and see whether
the stronger compensation becomes too aggressive.

This is still a single-frame recap step.
We are only building the candidate signals so that the next combos can compare them
numerically and visually.

```python
# Small epsilon for numerical safety, though ranges here should already be > 0.
eps = 1e-6

pseudo_reflectivity_r1 = intensity * np.maximum(ranges, eps)
pseudo_reflectivity_r2 = intensity * np.maximum(ranges, eps) ** 2

print("Pseudo-reflectivity features created.")
print("r1 shape:", pseudo_reflectivity_r1.shape)
print("r2 shape:", pseudo_reflectivity_r2.shape)

print("\nFirst 10 raw intensity values:")
print(intensity[:10])

print("\nFirst 10 pseudo-reflectivity values: I * R")
print(pseudo_reflectivity_r1[:10])

print("\nFirst 10 pseudo-reflectivity values: I * R^2")
print(pseudo_reflectivity_r2[:10])
```

## Compare the three signals numerically

We will summarize the distributions of the three single-frame signals:

- raw intensity,
- pseudo-reflectivity $I \cdot R$,
- pseudo-reflectivity $I \cdot R^2$.

Before plotting anything, we first inspect the scale and spread numerically.

This helps us answer a basic question:
how strongly do these range-aware transformations reshape the signal distribution?

If the stronger compensation grows too aggressively, we should see that directly in the
summary statistics through a much heavier upper tail and a much larger spread.

```python
def describe_signal(name, values):
    q = np.percentile(values, [0, 25, 50, 75, 95, 99, 100])
    print(f"{name}")
    print(f"  min   : {q[0]:.6f}")
    print(f"  p25   : {q[1]:.6f}")
    print(f"  p50   : {q[2]:.6f}")
    print(f"  p75   : {q[3]:.6f}")
    print(f"  p95   : {q[4]:.6f}")
    print(f"  p99   : {q[5]:.6f}")
    print(f"  max   : {q[6]:.6f}")
    print(f"  mean  : {values.mean():.6f}")
    print(f"  std   : {values.std():.6f}")
    print()

describe_signal("Raw intensity", intensity)
describe_signal("Pseudo-reflectivity (I * R)", pseudo_reflectivity_r1)
describe_signal("Pseudo-reflectivity (I * R^2)", pseudo_reflectivity_r2)
```

## Measure how each signal relates to distance from the sensor

We will now check how strongly each signal correlates with range.

This gives us a quick numerical sense of whether the transformed signals remain strongly
distance-driven.

We compare:

- raw intensity vs range,
- pseudo-reflectivity $I \cdot R$ vs range,
- pseudo-reflectivity $I \cdot R^2$ vs range.

This is an important diagnostic step because the whole point of pseudo-reflectivity is not to
pretend geometry vanishes, but to understand how the signal changes when range is explicitly
brought into the feature.

If the stronger proxy overcompensates, that should show up here as an even stronger positive
relationship with range.

```python
corr_intensity_range = np.corrcoef(intensity, ranges)[0, 1]
corr_r1_range = np.corrcoef(pseudo_reflectivity_r1, ranges)[0, 1]
corr_r2_range = np.corrcoef(pseudo_reflectivity_r2, ranges)[0, 1]

print(f"Correlation between raw intensity and range           : {corr_intensity_range:.6f}")
print(f"Correlation between pseudo-reflectivity I*R and range : {corr_r1_range:.6f}")
print(f"Correlation between pseudo-reflectivity I*R^2 and range: {corr_r2_range:.6f}")
```

We now compare the histograms of:

## Evaluate semantic usefulness numerically

We move from global signal distributions to semantic usefulness.

Global histograms tell us how the signal behaves overall, but they do not tell us whether
the signal helps distinguish semantic classes.

So here we do the numerical part first:

1. identify the valid semantic classes in this frame,
2. keep only classes with enough support,
3. compute a simple Fisher-style multiclass separability score,
4. compare raw intensity against pseudo-reflectivity $I \cdot R$.

This lets us test, on the same single frame, whether the range-aware proxy produces a more
semantically discriminative signal than raw intensity alone.

We are intentionally using $I \cdot R$ here, not $I \cdot R^2$, because the earlier
cells already showed that the stronger proxy looks numerically overcompensated.

```python
from collections import Counter

# Exclude unlabeled / ignored classes
ignore_labels = {0, 255}

# Count classes in this frame
class_counts = Counter(semantic_labels.tolist())

# Keep only classes with enough support
min_points = 500
valid_classes = sorted([
    cls for cls, cnt in class_counts.items()
    if cls not in ignore_labels and cnt >= min_points
])

print("Valid classes used for analysis:")
for cls in valid_classes:
    print(f"  Class {cls}: {class_counts[cls]} points")

def fisher_separability(signal, labels, classes):
    class_means = []
    class_vars = []
    class_ns = []

    for cls in classes:
        vals = signal[labels == cls]
        class_means.append(vals.mean())
        class_vars.append(vals.var())
        class_ns.append(len(vals))

    class_means = np.array(class_means)
    class_vars = np.array(class_vars)
    class_ns = np.array(class_ns)

    global_mean = np.average(class_means, weights=class_ns)
    between = np.sum(class_ns * (class_means - global_mean) ** 2)
    within = np.sum(class_ns * class_vars)

    score = between / (within + 1e-12)
    return score, class_means, class_vars, class_ns

score_intensity, means_I, vars_I, ns = fisher_separability(
    intensity, semantic_labels, valid_classes
)
score_r1, means_R1, vars_R1, _ = fisher_separability(
    pseudo_reflectivity_r1, semantic_labels, valid_classes
)

print("\nClass-separability score (higher is better):")
print(f"  Raw intensity           : {score_intensity:.8f}")
print(f"  Pseudo-reflectivity I*R : {score_r1:.8f}")

print("\nPer-class mean values:")
for i, cls in enumerate(valid_classes):
    print(
        f"  Class {cls:>3} | "
        f"mean(I) = {means_I[i]:.6f} | "
        f"mean(I*R) = {means_R1[i]:.6f} | "
        f"n = {ns[i]}"
    )
```

## Build FOV projections and summarize them with projection-space metrics

In this code, we project the LiDAR frame into a simple angular field-of-view image.

This gives us sensor-view visualizations of:

- range,
- raw intensity,
- pseudo-reflectivity $I \cdot R$,
- semantic labels.

Along with the images, we also compute projection-space metrics such as:

- pixel coverage,
- collision count,
- valid-pixel statistics.

This lets us check whether the projected signals remain structurally meaningful in sensor-view
space, not just in bird’s-eye view or in 1D histograms.

The goal here is not to build a production LiDAR projection model.
It is to create a compact, interpretable FOV representation that matches the preliminary
single-frame analysis.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1) Spherical projection
x = xyz[:, 0]
y = xyz[:, 1]
z = xyz[:, 2]

r = np.linalg.norm(xyz, axis=1) + 1e-6
yaw = np.arctan2(y, x)
pitch = np.arcsin(z / r)

W = 1024
H = 64

yaw_min, yaw_max = -np.pi, np.pi
pitch_min, pitch_max = pitch.min(), pitch.max()

u = ((yaw - yaw_min) / (yaw_max - yaw_min) * (W - 1)).astype(np.int32)
v = ((pitch_max - pitch) / (pitch_max - pitch_min) * (H - 1)).astype(np.int32)

u = np.clip(u, 0, W - 1)
v = np.clip(v, 0, H - 1)

# 2) Projection buffers
range_img = np.full((H, W), np.nan, dtype=np.float32)
intensity_img = np.full((H, W), np.nan, dtype=np.float32)
pseudo_r1_img = np.full((H, W), np.nan, dtype=np.float32)
semantic_img = np.full((H, W), np.nan, dtype=np.float32)

filled_mask = np.zeros((H, W), dtype=bool)
collision_count = 0

# Far-to-near so nearer points overwrite farther ones
order = np.argsort(r)[::-1]

for idx in order:
    row, col = v[idx], u[idx]
    if filled_mask[row, col]:
        collision_count += 1
    filled_mask[row, col] = True

    range_img[row, col] = r[idx]
    intensity_img[row, col] = intensity[idx]
    pseudo_r1_img[row, col] = pseudo_reflectivity_r1[idx]
    semantic_img[row, col] = semantic_labels[idx]

# 3) Projection-space metrics
total_pixels = H * W
valid_pixels = int(np.sum(filled_mask))
coverage_ratio = valid_pixels / total_pixels
points_projected = len(r)
collisions_ratio = collision_count / points_projected

def valid_stats(name, img):
    vals = img[~np.isnan(img)]
    print(name)
    print(f"  valid pixels : {len(vals)}")
    print(f"  min          : {vals.min():.6f}")
    print(f"  p25          : {np.percentile(vals, 25):.6f}")
    print(f"  p50          : {np.percentile(vals, 50):.6f}")
    print(f"  p75          : {np.percentile(vals, 75):.6f}")
    print(f"  p95          : {np.percentile(vals, 95):.6f}")
    print(f"  max          : {vals.max():.6f}")
    print(f"  mean         : {vals.mean():.6f}")
    print(f"  std          : {vals.std():.6f}")
    print()

projected_classes = np.unique(semantic_img[~np.isnan(semantic_img)]).astype(np.int32)

print("Projection-space metrics")
print(f"  Image size                : {H} x {W} = {total_pixels} pixels")
print(f"  Total input points        : {points_projected}")
print(f"  Valid projected pixels    : {valid_pixels}")
print(f"  Pixel coverage ratio      : {coverage_ratio:.6f}")
print(f"  Pixel collision count     : {collision_count}")
print(f"  Pixel collision ratio     : {collisions_ratio:.6f}")
print(f"  Unique projected classes  : {len(projected_classes)}")
print(f"  Projected class IDs       : {projected_classes.tolist()}")
print()

valid_stats("Projected range image stats", range_img)
valid_stats("Projected raw intensity image stats", intensity_img)
valid_stats("Projected pseudo-reflectivity (I * R) image stats", pseudo_r1_img)

# 4) Visualizations
plt.figure(figsize=(14, 3))
plt.imshow(range_img, aspect="auto")
plt.title("FOV Projection: Range")
plt.xlabel("Horizontal angle bins")
plt.ylabel("Vertical angle bins")
plt.colorbar()
plt.show()

plt.figure(figsize=(14, 3))
plt.imshow(intensity_img, aspect="auto")
plt.title("FOV Projection: Raw Intensity")
plt.xlabel("Horizontal angle bins")
plt.ylabel("Vertical angle bins")
plt.colorbar()
plt.show()

plt.figure(figsize=(14, 3))
plt.imshow(pseudo_r1_img, aspect="auto")
plt.title("FOV Projection: Pseudo-Reflectivity (I * R)")
plt.xlabel("Horizontal angle bins")
plt.ylabel("Vertical angle bins")
plt.colorbar()
plt.show()

plt.figure(figsize=(14, 3))
plt.imshow(semantic_img, aspect="auto")
plt.title("FOV Projection: Semantic Labels")
plt.xlabel("Horizontal angle bins")
plt.ylabel("Vertical angle bins")
plt.colorbar()
plt.show()
```

- load the semantic labels,
- keep only sufficiently supported semantic classes,
- compute the separability score for raw intensity,
- compute the separability score for pseudo-reflectivity $I \cdot R$,
- and record the frame-wise gain.

This gives us a direct temporal-semantic comparison across the full motion window.

```python
def load_semantickitti_labels(label_path):
    raw = np.fromfile(label_path, dtype=np.uint32)
    semantic = raw & 0xFFFF
    return semantic

def get_valid_classes_notebook02_style(labels, min_points_per_class=500, ignored_classes=(0, 1, 255)):
    valid_classes = []
    unique_classes, counts = np.unique(labels, return_counts=True)

    for cls, count in zip(unique_classes, counts):
        if cls in ignored_classes:
            continue
        if count >= min_points_per_class:
            valid_classes.append(int(cls))

    return sorted(valid_classes)

def fisher_separability(values, labels, valid_classes):
    class_means = []
    class_vars = []
    class_ns = []

    for cls in valid_classes:
        cls_values = values[labels == cls]
        class_means.append(np.mean(cls_values))
        class_vars.append(np.var(cls_values))
        class_ns.append(len(cls_values))

    class_means = np.array(class_means, dtype=np.float64)
    class_vars = np.array(class_vars, dtype=np.float64)
    class_ns = np.array(class_ns, dtype=np.float64)

    global_mean = np.sum(class_ns * class_means) / np.sum(class_ns)
    between = np.sum(class_ns * (class_means - global_mean) ** 2)
    within = np.sum(class_ns * class_vars)

    score = between / (within + 1e-12)
    return float(score), class_means, class_vars, class_ns

semantic_rows = []

for frame in multi_frame_data:
    fid = frame["frame_id"]
    label_path = LABELS_DIR / f"{fid}.label"
    semantic_labels_frame = load_semantickitti_labels(label_path)

    valid_classes_frame = get_valid_classes_notebook02_style(
        semantic_labels_frame,
        min_points_per_class=500,
        ignored_classes=(0, 1, 255)
    )

    score_intensity_frame, _, _, ns = fisher_separability(
        frame["intensity"],
        semantic_labels_frame,
        valid_classes_frame
    )

    score_pr_frame, _, _, _ = fisher_separability(
        frame["pseudo_reflectivity"],
        semantic_labels_frame,
        valid_classes_frame
    )

    semantic_rows.append({
        "frame_id": fid,
        "num_valid_classes": len(valid_classes_frame),
        "valid_classes": valid_classes_frame,
        "intensity_sep_score": score_intensity_frame,
        "pr_sep_score": score_pr_frame,
        "absolute_gain": score_pr_frame - score_intensity_frame,
        "relative_gain": (score_pr_frame - score_intensity_frame) / (score_intensity_frame + 1e-12),
    })

semantic_df = pd.DataFrame(semantic_rows)

print("Frame 000000 sanity check")
row0 = semantic_df.iloc[0]
print(" - valid classes       :", row0["valid_classes"])
print(" - num valid classes   :", row0["num_valid_classes"])
print(" - intensity score     :", f'{row0["intensity_sep_score"]:.6f}')
print(" - pseudo-reflectivity :", f'{row0["pr_sep_score"]:.6f}')
print(" - absolute gain       :", f'{row0["absolute_gain"]:.6f}')
print(" - relative gain       :", f'{row0["relative_gain"]:.6f}')
print()

print("First 10 frame-wise results")
print(semantic_df[[
    "frame_id",
    "num_valid_classes",
    "intensity_sep_score",
    "pr_sep_score",
    "absolute_gain",
    "relative_gain"
]].head(10))
print()

print("30-frame semantic separability summary")
print(" - intensity score mean        :", semantic_df["intensity_sep_score"].mean())
print(" - pseudo-reflectivity mean    :", semantic_df["pr_sep_score"].mean())
print(" - absolute gain mean          :", semantic_df["absolute_gain"].mean())
print(" - relative gain mean          :", semantic_df["relative_gain"].mean())
```

## Quantifying semantic separability across valid classes

In this code, we measure directly whether pseudo-reflectivity separates the valid semantic
classes more clearly than raw intensity on this single frame.

We use a Fisher-style multiclass separability score across the valid semantic classes.

This gives one compact number for:

- raw intensity,
- pseudo-reflectivity $I \cdot R$,

and lets us compare the class-wise semantic usefulness numerically before moving on.

```python
import numpy as np

def fisher_style_multiclass_score(values, labels, classes):
    overall_mean = np.mean(values)

    between = 0.0
    within = 0.0

    for cls in classes:
        cls_vals = values[labels == cls]
        cls_mean = np.mean(cls_vals)
        cls_var = np.var(cls_vals)
        n_cls = len(cls_vals)

        between += n_cls * (cls_mean - overall_mean) ** 2
        within += n_cls * cls_var

    return between / within if within > 0 else np.nan

intensity_score_sem = fisher_style_multiclass_score(
    intensity_sem, semantic_labels_sem, valid_classes_semantic
)
pseudo_score_sem = fisher_style_multiclass_score(
    pseudo_reflectivity_sem, semantic_labels_sem, valid_classes_semantic
)

print("Multiclass semantic separability score")
print(f"Raw intensity          : {intensity_score_sem:.6f}")
print(f"Pseudo-reflectivity    : {pseudo_score_sem:.6f}")
print(f"Absolute improvement   : {pseudo_score_sem - intensity_score_sem:.6f}")
print(f"Relative improvement   : {(pseudo_score_sem - intensity_score_sem) / intensity_score_sem:.6f}")
```

For each retained class in each frame, we record:

- point count,
- mean and standard deviation of raw intensity,
- mean and standard deviation of pseudo-reflectivity,
- mean and standard deviation of range.

This creates the temporal semantic table that the final recap section is really about.

Once this table exists, we can ask the key question:
do the class-wise patterns from the single-frame semantic recap persist across motion?

```python
MIN_POINTS_PER_CLASS_05 = 200

class_rows_05 = []

for frame_id in frame_ids_05:
    xyz_05, intensity_05 = load_frame_points_05(frame_id)
    semantic_labels_05 = load_frame_labels_05(frame_id)
    ranges_05, pseudo_reflectivity_05 = compute_range_and_pseudo_reflectivity_05(xyz_05, intensity_05)

    unique_classes_05, counts_05 = np.unique(semantic_labels_05, return_counts=True)

    for class_id, count in zip(unique_classes_05, counts_05):
        if count < MIN_POINTS_PER_CLASS_05:
            continue

        mask = semantic_labels_05 == class_id

        class_rows_05.append({
            "frame_id": frame_id,
            "class_id": int(class_id),
            "num_points": int(count),
            "intensity_mean": float(np.mean(intensity_05[mask])),
            "intensity_std": float(np.std(intensity_05[mask])),
            "pseudo_reflectivity_mean": float(np.mean(pseudo_reflectivity_05[mask])),
            "pseudo_reflectivity_std": float(np.std(pseudo_reflectivity_05[mask])),
            "range_mean": float(np.mean(ranges_05[mask])),
            "range_std": float(np.std(ranges_05[mask])),
        })

class_summary_df_05 = (
    pd.DataFrame(class_rows_05)
    .sort_values(by=["frame_id", "class_id"])
    .reset_index(drop=True)
)

print("MIN_POINTS_PER_CLASS:", MIN_POINTS_PER_CLASS_05)
print("Number of retained class-frame rows:", len(class_summary_df_05))
print("Unique retained classes:", sorted(class_summary_df_05["class_id"].unique().tolist()))
print()

class_summary_df_05.head(20)
```

## Keeping only temporally persistent classes

In this part, we keep only those semantic classes that survive across all selected frames.

This gives us a clean basis for short-window semantic consistency analysis.

The reason for doing this is simple:

if a class appears only in some frames, then its temporal behavior is harder to compare fairly
across the full window.

So before measuring class-wise stability or class-order consistency, we first restrict the
analysis to classes that are present in all 10 selected frames.

```python
num_selected_frames_05 = len(frame_ids_05)

class_frame_counts_05 = (
    class_summary_df_05.groupby("class_id")["frame_id"]
    .nunique()
    .reset_index(name="num_frames_present")
)

persistent_classes_05 = class_frame_counts_05.loc[
    class_frame_counts_05["num_frames_present"] == num_selected_frames_05,
    "class_id"
].tolist()

persistent_summary_df_05 = class_summary_df_05[
    class_summary_df_05["class_id"].isin(persistent_classes_05)
].copy().reset_index(drop=True)

print("Selected frames:", num_selected_frames_05)
print("Persistent classes:", persistent_classes_05)
print("Number of persistent classes:", len(persistent_classes_05))
print("Persistent summary shape:", persistent_summary_df_05.shape)
print()

class_frame_counts_05.sort_values(
    ["num_frames_present", "class_id"],
    ascending=[False, True]
)
```

## Measuring class-wise temporal stability

In this part, we quantify how much each persistent class mean changes across the selected
10-frame motion window for both:

- raw intensity,
- pseudo-reflectivity.

This gives a first direct measure of short-horizon semantic consistency.

For each persistent class, we compute:

- mean value over time,
- standard deviation over time,
- coefficient of variation over time.

The goal is not to prove invariance.
It is to check whether the class-wise signal behavior remains reasonably stable across motion.

```python
stability_rows_05 = []

for class_id in persistent_classes_05:
    cls_df = persistent_summary_df_05[
        persistent_summary_df_05["class_id"] == class_id
    ].sort_values("frame_id")

    intensity_means = cls_df["intensity_mean"].values
    pseudo_means = cls_df["pseudo_reflectivity_mean"].values

    stability_rows_05.append({
        "class_id": int(class_id),
        "num_frames": len(cls_df),
        "intensity_mean_over_time": float(np.mean(intensity_means)),
        "intensity_std_over_time": float(np.std(intensity_means)),
        "intensity_cv_over_time": float(np.std(intensity_means) / (np.mean(intensity_means) + 1e-12)),
        "pseudo_mean_over_time": float(np.mean(pseudo_means)),
        "pseudo_std_over_time": float(np.std(pseudo_means)),
        "pseudo_cv_over_time": float(np.std(pseudo_means) / (np.mean(pseudo_means) + 1e-12)),
    })

stability_df_05 = (
    pd.DataFrame(stability_rows_05)
    .sort_values("class_id")
    .reset_index(drop=True)
)

stability_df_05
```

## Test whether class ordering persists across frames

In this code, we test whether the class-wise ordering remains stable across the 10-frame
motion window.

Instead of asking whether every class mean stays numerically identical, we ask a more robust
question:

does the relative ordering of classes remain similar from frame to frame?

To measure this, we:

- build a frame-by-class table for raw intensity means,
- build a frame-by-class table for pseudo-reflectivity means,
- compute a reference class profile by averaging across the full window,
- and then compute the Spearman rank correlation between each frame and that reference profile.

This gives a compact measure of short-window semantic consistency through motion.

```python
from scipy.stats import spearmanr

# Pivot tables: rows = frame_id, columns = class_id
intensity_pivot_05 = persistent_summary_df_05.pivot(
    index="frame_id", columns="class_id", values="intensity_mean"
).sort_index()

pseudo_pivot_05 = persistent_summary_df_05.pivot(
    index="frame_id", columns="class_id", values="pseudo_reflectivity_mean"
).sort_index()

# Reference profile = average class means across the full frame window
intensity_reference_05 = intensity_pivot_05.mean(axis=0)
pseudo_reference_05 = pseudo_pivot_05.mean(axis=0)

ordering_rows_05 = []

for frame_id in intensity_pivot_05.index:
    intensity_corr, _ = spearmanr(intensity_pivot_05.loc[frame_id], intensity_reference_05)
    pseudo_corr, _ = spearmanr(pseudo_pivot_05.loc[frame_id], pseudo_reference_05)

    ordering_rows_05.append({
        "frame_id": frame_id,
        "intensity_rank_corr_vs_mean_profile": float(intensity_corr),
        "pseudo_rank_corr_vs_mean_profile": float(pseudo_corr),
    })

ordering_df_05 = pd.DataFrame(ordering_rows_05)

print("Class-ordering consistency across frames")
display(ordering_df_05)

print("\nAverage rank correlation")
print("Intensity           :", round(ordering_df_05["intensity_rank_corr_vs_mean_profile"].mean(), 4))
print("Pseudo-reflectivity :", round(ordering_df_05["pseudo_rank_corr_vs_mean_profile"].mean(), 4))
```

## Visualizing ordering consistency across motion

In this code, we plot the frame-wise rank-correlation values for:

- raw intensity,
- pseudo-reflectivity.

This gives a compact visual check of how stable the semantic class ordering remains as the
platform moves.

The goal is to make the consistency result easier to read across the 10-frame window and to
see whether the pseudo-reflectivity advantage is steady, weak, or erratic.

```python
plt.figure(figsize=(8, 4))

plt.plot(
    ordering_df_05["frame_id"],
    ordering_df_05["intensity_rank_corr_vs_mean_profile"],
    marker="o",
    label="Raw intensity"
)

plt.plot(
    ordering_df_05["frame_id"],
    ordering_df_05["pseudo_rank_corr_vs_mean_profile"],
    marker="o",
    label="Pseudo-reflectivity"
)

plt.ylim(0.85, 1.00)
plt.xlabel("Frame ID")
plt.ylabel("Spearman rank correlation")
plt.title("Class-ordering consistency across motion")
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

## Summarizing the semantic-consistency result numerically

This gives us a neat final quantitative statement for the short-window semantic-consistency
recap without bloating the analysis.

The goal is to summarize:

- average rank consistency,
- minimum and maximum consistency,
- how often pseudo-reflectivity is better,
- how often raw intensity is better,
- and the mean advantage of pseudo-reflectivity over raw intensity.

```python
intensity_vals_05 = ordering_df_05["intensity_rank_corr_vs_mean_profile"].values
pseudo_vals_05 = ordering_df_05["pseudo_rank_corr_vs_mean_profile"].values

summary_05 = {
    "num_frames": len(ordering_df_05),
    "intensity_mean_rank_corr": float(np.mean(intensity_vals_05)),
    "pseudo_mean_rank_corr": float(np.mean(pseudo_vals_05)),
    "intensity_min_rank_corr": float(np.min(intensity_vals_05)),
    "pseudo_min_rank_corr": float(np.min(pseudo_vals_05)),
    "intensity_max_rank_corr": float(np.max(intensity_vals_05)),
    "pseudo_max_rank_corr": float(np.max(pseudo_vals_05)),
    "pseudo_better_frames": int(np.sum(pseudo_vals_05 > intensity_vals_05)),
    "intensity_better_frames": int(np.sum(intensity_vals_05 > pseudo_vals_05)),
    "tied_frames": int(np.sum(np.isclose(pseudo_vals_05, intensity_vals_05))),
    "mean_advantage_pseudo_minus_intensity": float(np.mean(pseudo_vals_05 - intensity_vals_05)),
}

for k, v in summary_05.items():
    if isinstance(v, float):
        print(f"{k}: {v:.4f}")
    else:
        print(f"{k}: {v}")
```
